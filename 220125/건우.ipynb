## 스택 
https://level.goorm.io/exam/43218/%EC%8A%A4%ED%83%9D-stack/quiz/1

```python
count = int(input())

result = []
status = 0
for i in range(count):
    flag = int(input())
    if flag == 0:
        push = input()
        result.append(push)
    elif flag == 1 and len(result) == 0:
        print('underflow')
        status = 1
    elif flag == 0 and len(result) > 10:
        print('overflow')
        status = 1
    elif flag == 1:
        result.pop()
    else:
        print(' '.join(result), end=' ')
        break
print(' '.join(result), end=' ')
```
