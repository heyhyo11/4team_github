# [Groom] 스택(Stack)
https://level.goorm.io/exam/43218/%EC%8A%A4%ED%83%9D-stack/quiz/1

```python
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
user_input = int(input())
stack = []

for i in range(user_input):
	type_input = int(input())
	if type_input == 0:
			data_input = input()
			stack.append(data_input)
	elif type_input == 1:
		if len(stack) == 0:
			print('underflow')
		else:
			del stack[len(stack)-1]
	else:
		break


if len(stack) > 10:
	print('overflow', end= ' ')

for i in range(len(stack[0:10])):
	print(stack[i], end = ' ')
```

# [프로그래머스] 코딩테스트 연습 > 연습문제 > 핸드폰 번호 가리기
https://programmers.co.kr/learn/courses/30/lessons/12948

```python
def solution(phone_number):
    back_number = phone_number[-4:]
    front_number = phone_number[0:-4]
    front_number = '*' * len(front_number)
    full_number = front_number + back_number
    return full_number
```

```python
# 다른 사람의 풀이
def hide_numbers(s):
    return "*"*(len(s)-4) + s[-4:]
```
